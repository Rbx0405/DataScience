```python
#joining of dataframe
import pandas as pd
import numpy as np
df1= pd.DataFrame({'id':['A01','A02','A03','A04'],'Name':['ABC','DEF','GHI','JKL']})
df2= pd.DataFrame({'id':['B01','B02','B03','B04'],'Name':['CBA','FED','IHG','LJK']})
result = pd.concat([df1, df2],ignore_index=True)

display(result)
```

```python
#filtering
import pandas as pd
import numpy as np
df1= pd.DataFrame({'id':['A01','A02','A03','A04'],'Name':['ABC','DEF','GHI','JKL']})
df2= pd.DataFrame({'id':['B01','B02','B03','B04'],'Name':['CBA','FED','IHG','LJK']})
result = pd.concat([df1, df2])
filtered = result[result['Name'] == 'ABC']
display(filtered)
```

```python
#Age Distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df["Age"].plot(kind="hist", bins=10, alpha=0.7, color="skyblue", density=True)
df["Age"].plot(kind="kde", color="darkblue", lw=2)  # smooth curve
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show
```

```python
#Salary Distribution
plt.figure(figsize=(8,5))
df["Salary"].dropna().plot(kind="hist", bins=20, alpha=0.7, color="lightgreen", density=True)
df["Salary"].dropna().plot(kind="kde", color="darkgreen", lw=2)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Density")
plt.show()
```
